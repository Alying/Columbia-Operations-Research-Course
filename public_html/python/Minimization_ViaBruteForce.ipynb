```python
import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D


# Plot the surface

# f = 2*(X-a).^2 + 5*(Y-b).^2 + (X-a).*(Y-b) + c;
def fEval(x, y):
    return 2*(x-20.23)**2 + 5*(y+31.23)**2 + (x-20.23)*(y+31.23) + 10

X = np.linspace(-200,200, 100)
Y = np.linspace(-200,200, 100)
X, Y =np.meshgrid(X,Y)

Z=fEval(X,Y)

fig=plt.figure()
ax=fig.gca(projection='3d')
surf=ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, linewidth=0, antialiased=False)
fig.colorbar(surf, shrink=0.5, aspect=5)

# Grid Search --- Brute_Force 

fMin = 1.0e18;
for x in np.arange (-40, 40, 0.5):
    for y in np.arange (-40, 40, 0.5):
        f = fEval(x,y)
        if f < fMin:
            fMin = f
            xX = x
            yY = y
print ("min point: x={},y={} and Z={}".format(xX, yY, fMin))
plt.show()
```
