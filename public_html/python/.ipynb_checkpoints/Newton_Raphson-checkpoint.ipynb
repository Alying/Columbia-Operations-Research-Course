```python
import math
import numpy as np
import matplotlib.pyplot as plt

tol=0.0000001
count=0

x=4
a=13
uBound=30
lBound=-uBound

xX=np.arange(lBound, uBound, 0.01)
f=xX**2-a
y=0*xX
plt.plot(xX, f)
plt.plot(xX, y)
plt.show()

count=0
while (1):
    xPre=x
    x=x-(x**2-a)/(2.0*x)
    count+=1
    if (abs(x-xPre)<tol): break

print("Count: {} Sqrt(13)= {} vs. Newton-Raphson Solution: {}".format(count, math.sqrt(13), x))
```
